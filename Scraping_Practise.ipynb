```python
import requests
from bs4 import BeautifulSoup
import lxml
import pandas as pd
```

```python
url = 'https://www.bolton.ac.uk/subject-areas'
# url_2 = 'https://play.google.com/store/search?q=education&c=apps'
response = requests.get(url)
html_content = response.text
```

```python
html_content
```

```python
soup = BeautifulSoup(html_content, 'lxml')

# top_free_apps_names = soup.find_all(name='span', class_='wMUdtb')
# top_free_apps_rating = soup.find_all(name='span', class_='w2kbF')
# top_free_apps_heading = soup.find_all(name='span', class_='DdYX5')

subject_areas = soup.find_all(name='div', class_='LinkBlocktitle')

subject_areas_storage = []

# app_names = []
# app_rating = []
# app_heading = []

for subject in subject_areas:
  subject_areas_storage.append(subject.text)

# for app in top_free_apps_names:
#   app_names.append(''.join(app))

# for rate in top_free_apps_rating:
#   app_rating.append(float(''.join(rate)))

# for head in top_free_apps_heading:
#   app_heading.append(''.join(head))
```

```python
print(subject_areas_storage)
```

```python
bolton_courses_scrape = pd.DataFrame({
    'Subject Area': subject_areas_storage,
})
```
